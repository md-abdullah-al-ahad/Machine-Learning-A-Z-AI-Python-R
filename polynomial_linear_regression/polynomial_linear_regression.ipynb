```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Loading the dataset
dataset = pd.read_csv('Position_Salaries.csv')
```

```python
# Separating features (X) and target variable (y)
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
# Training the Linear Regression model for comparison
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
# Training the Polynomial Regression model
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

```python
#Plotting the linear regression model
sns.set_style("whitegrid")
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='red', edgecolors='black', s=80, alpha=0.7, label='Data Points')
x_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
plt.plot(x_range, lin_reg.predict(x_range), color='blue', linewidth=2.5, linestyle='--', label='Regression Line')
plt.title('Linear Regression', fontsize=14, fontweight='bold', color='darkblue')
plt.xlabel('Position Level', fontsize=12, color='darkgreen')
plt.ylabel('Salary', fontsize=12, color='darkgreen')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

```python
#Plotting the Polynomial regression model
sns.set_style("whitegrid")
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='red', edgecolors='black', s=80, alpha=0.7, label='Data Points')
x_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
plt.plot(x_range, lin_reg_2.predict(poly_reg.fit_transform(x_range)), color='blue', linewidth=2.5, linestyle='--', label='Polynomial Regression Line')
plt.title('Polynomial Regression', fontsize=14, fontweight='bold', color='darkblue')
plt.xlabel('Position Level', fontsize=12, color='darkgreen')
plt.ylabel('Salary', fontsize=12, color='darkgreen')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

```python
# Predicting the salary of a person in level 4.5 with linear regression
lin_reg.predict([[4.5]])
```

```python
# Predicting the salary of a person in level 4.5 with polynomial regression
lin_reg_2.predict(poly_reg.fit_transform([[4.5]]))
```
