```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Loading the dataset
dataset = pd.read_csv('Salary_Data.csv')
# Separating features (X) and target variable (y)
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
# Splitting data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Training the Linear Regression model
regr = LinearRegression()
regr.fit(X_train, y_train)
```

```python
# Predicting on test data
y_pred = regr.predict(X_test)
```

```python
# Plotting training data
sns.set_style("whitegrid")
plt.figure(figsize=(8, 5))
plt.scatter(X_train, y_train, color='red', edgecolors='black', s=80, alpha=0.7, label='Training Data')
x_range = np.linspace(X_train.min(), X_train.max(), 100).reshape(-1, 1)
plt.plot(x_range, regr.predict(x_range), color='blue', linewidth=2.5, linestyle='--', label='Regression Line')
plt.title('Salary vs Experience (Training set)', fontsize=14, fontweight='bold', color='darkblue')
plt.xlabel('Years of Experience', fontsize=12, color='darkgreen')
plt.ylabel('Salary', fontsize=12, color='darkgreen')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

```python
# Plotting test data
sns.set_style("whitegrid")
plt.figure(figsize=(8, 5))
plt.scatter(X_test, y_test, color='red', edgecolors='black', s=80, alpha=0.7, label='Test Data')
x_range = np.linspace(X_train.min(), X_train.max(), 100).reshape(-1, 1)
plt.plot(x_range, regr.predict(x_range), color='blue', linewidth=2.5, linestyle='--', label='Regression Line')
plt.title('Salary vs Experience (Test set)', fontsize=14, fontweight='bold', color='darkblue')
plt.xlabel('Years of Experience', fontsize=12, color='darkgreen')
plt.ylabel('Salary', fontsize=12, color='darkgreen')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

```python
# Model_performence
print(regr.coef_)
print(regr.intercept_)
accuracy = regr.score(X_test, y_test)
print(accuracy*100)
```
